```python
import warnings
warnings.filterwarnings('ignore')

from nilearn import datasets
import numpy as np
import warnings
from nilearn import plotting
from nilearn import image
from nilearn.datasets import load_mni152_template
import matplotlib.pyplot as plt

task_label = ['N-back', 'GradCPT', 'SST', 'Card', 'Eyes', 'Movies', 'Rest1', 'Rest2']
```

## State 1: N-back Task

```python
task = task_label[0]
img = image.load_img(str('Parcellations/Maj_'+task+'.nii'))
plotting.view_img(img, bg_img=None, draw_cross=False, colorbar=False, black_bg=True, 
              cmap = plt.cm.gist_ncar, symmetric_cmap=False, title=str(task+ ' Task'))
```
